```python
Titles = []
Dates = []
Categorys = []
Authors = []
Textos = []

for blog in Blog_containers:

  Title = blog.h3.a.text
  Titles.append(Title)

  date = blog.div.find('a', class_ = 'data-link')
  Dates.append(date)

  category = blog.div.find('span', class_ = 'category-link')
  Categorys.append(first_category.a.text)

  autor = blog.div.find('a', class_ = 'author vcard')
  Authors.append(autor.span.text)

  texto = blog.p.text
  Textos.append(texto)

import pandas as pd
test_df = pd.DataFrame({'Titulos': Titles,
                        'Fechas': Dates,
                        'Categorias': Categorys,
                        'Autores': Authors,
                        'Resumen' :Textos
                        })
print(test_df.info())
test_df
```

```python
import pandas as pd
Titles_ = []
Dates_ = []
Categorys_ = []
Authors_ = []
Textos_ = []
start_time = time()
requests = 0
pages = [str(i) for i in range(1,4)]

for page in pages:
  url = 'https://mascotascol.com/blog/page/'+ page + '/'
  print(url)
  response = get(url)

  requests += 1
  elapsed_time = time() - start_time
  print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
  clear_output(wait = True)

  page_html = BeautifulSoup(response.text, 'html.parser')
  Blg_containers = page_html.find_all('div', class_ = 'blog-content wf-td')

  for blog in Blg_containers:

    Title = blog.h3.a.text
    Titles_.append(Title)

    date = blog.div.find('a', class_ = 'data-link')
    Dates_.append(date)

    category = blog.div.find('span', class_ = 'category-link')
    Categorys_.append(first_category.a.text)

    autor = blog.div.find('a', class_ = 'author vcard')
    Authors_.append(autor.span.text)

    texto = blog.p.text
    Textos_.append(texto)

Final_df = pd.DataFrame({'Titulos': Titles_,
                        'Fechas': Dates_,
                        'Categorias': Categorys_,
                        'Autores': Authors_,
                        'Resumen' :Textos_
                        })
analisis_df = pd.DataFrame({'Titulos': Titles,
                        'Fechas': Dates,
                        'Categorias': Categorys,
                        'Autores': Authors,
                        'Resumen' :Textos
                        })
#print(test_df.info())
Final_df
```
